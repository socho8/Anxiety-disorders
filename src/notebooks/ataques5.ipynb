```python
#FUNCION QUE SUMA GENTE A LA LISTA QUE SUFRE ATAQUES DE PANICO YA QUE ARRIBA NO ESTA TOTALMENTE CORRECTO
def cambiar_panic(ataques):
    antecedentes = ataques['Antecedentes familiares']
    sintomas = ataques['Sintomas']

    if sintomas == 'Ataques de panico':
        return 1
    else:
        return 0
    
ataques['Panic Disorder Diagnosis'] = ataques.apply(cambiar_panic, axis = 1)
```

```python
#CHEQUEAMOS DE QUE LA FUNCION FUNCIONO CORRECTAMENTE
ataques[ataques['Panic Disorder Diagnosis'] == 1]
```

```python
#CREAMOS UNA FUNCION PARA CREAR UNA NUEVA COLUMNA CON LOS DIFERENTES NIVELES DE ANSIEDAD Y QUE QUEDE MASOMENOS BALANCEADA
def asignar_nivel(ataques):
    antecedentes = ataques["Antecedentes familiares"]
    estres = ataques['Nivel de estres']
    sintomas = ataques['Sintomas']
    severidad = ataques['Severidad']
    impacto = ataques['Impacto en la vida']
    apoyo = ataques['Apoyo social']
    diagnostico = ataques['Panic Disorder Diagnosis']

    # Asigna el nivel de ataques basado en las condiciones dadas
    if diagnostico == 1:
            return 4
    elif diagnostico == 0:
        if sintomas in ['Ataques de panico']:
            return 4
        elif sintomas in ['Dificultad para respirar'] and antecedentes == 'Si' and apoyo == 'Bajo':
            return 4
        elif sintomas in ['Dificultad para respirar'] and antecedentes == 'Si' and apoyo in ['Moderado', 'Alto']:
            return 3
        elif sintomas in ['Dificultad para respirar'] and antecedentes == 'No':
            return 3
        elif sintomas in ['Dolor de pecho'] and antecedentes == 'Si' and apoyo == 'Bajo':
            return 3
        elif sintomas in ['Dolor de pecho '] and antecedentes == 'Si' and apoyo in ['Moderado', 'Alto']:
            if severidad == 'Alto':
                return 3
            else:
                return 2
        elif sintomas in ['Dolor de pecho '] and antecedentes == 'No':
            if severidad == 'Alto':
                return 3
            else: 
                return 2
        elif sintomas in ['Miedo a perder el control'] and antecedentes == 'Si' and apoyo == 'Bajo':
            return 3
        elif sintomas in ['Miedo a perder el control'] and antecedentes == 'Si' and apoyo in ['Moderado', 'Alto']:
            if severidad == 'Alto':
                return 3
            else:
                return 2
        elif sintomas in ['Miedo a perder el control'] and antecedentes == 'No':
            return 2
        elif sintomas in ['Mareos'] and antecedentes == 'Si' and apoyo in ['Bajo', 'Moderado']:
            return 2
        elif sintomas in ['Mareos'] and antecedentes == 'No' and apoyo in ['Alto']:
            if severidad in ['Alto', 'Moderado']:
                return 2
            else:
                return 1
        else:
            return 1
    else:
        return None  # Si no cumple ninguna de las condiciones, puede devolver un valor predeterminado o None
    

ataques['Nivel de ansiedad'] = ataques.apply(asignar_nivel, axis=1)
```

```python
#CAMBIAMOS LA DATA A VALORES NUMERICOS
genero_num = {'Hombre' : 0 , 'Mujer' : 1}
antecedentes_num = {'No' : 0, 'Si' : 1}
estres_num = {'Bajo' : 0 , 'Moderado' : 1 , 'Alto' : 2}
sintomas_num = {'Mareos' : 0, 'Miedo a perder el control' : 1, 'Dolor de pecho' : 2, 'Dificultad para respirar':3,'Ataques de panico' : 4}
severidad_num = {'Bajo' : 0 , 'Moderado' : 1 , 'Alto' : 2}
impacto_num = {'Bajo' : 0 , 'Moderado' : 1 , 'Alto' : 2}
apoyo_num = {'Bajo' : 0 , 'Moderado' : 1 , 'Alto' : 2}

ataques["Genero"] = ataques["Genero"].replace(genero_num)
ataques["Antecedentes familiares"] = ataques["Antecedentes familiares"].replace(antecedentes_num)
ataques["Nivel de estres"] = ataques["Nivel de estres"].replace(estres_num)
ataques["Sintomas"] = ataques["Sintomas"].replace(sintomas_num)
ataques["Impacto en la vida"] = ataques["Impacto en la vida"].replace(impacto_num)
ataques["Severidad"] = ataques["Severidad"].replace(severidad_num)
ataques["Apoyo social"] = ataques["Apoyo social"].replace(apoyo_num)
```

```python
#GUARDAMOS CSV FINAL CON LA DATA NUMERICA Y YA MODIFICADA
ataques.to_csv('Ansiedad_final_num.csv')
```

```python
#SEPARAMOS LA DATA PARA SU ENTRENAMIENTO
X = ataques.drop('Nivel de ansiedad', axis=1)
y = ataques['Nivel de ansiedad']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

```python
#IMPORTAMOS EL MODELO
from sklearn.ensemble import RandomForestClassifier
```

```python
#UTILIZAREMOS UN MODELO DE RANDOM FOREST CLASSIFIER
rnd_clf = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
#ENTRENAMOS EL MODELO
rnd_clf.fit(X_train, y_train)
```

```python
#PREDECIMOS
y_pred_RFC = rnd_clf.predict(X_test)
```

```python
#HACEMOS LA MATRIZ DE CONFUSION
from sklearn.metrics import confusion_matrix, accuracy_score

conf_matrixRFC = confusion_matrix(y_test, y_pred_RFC)

# Imprimir la matriz de confusiÃ³n
print("Matriz de ConfusiÃ³n:")
print(conf_matrixRFC)
```

```python
#LA MOSTRAMOS CON UN GRAFICO
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

plt.figure(figsize=(10, 10))
sns.heatmap(conf_matrixRFC, annot=True, fmt='d', cmap='Blues')

plt.title('Matriz de ConfusiÃ³n')
plt.xlabel('Predicciones')
plt.ylabel('Valores Reales')
plt.show()
```

```python
#VEMOS SU PRECISION
accuracyConf = accuracy_score(y_test, y_pred_RFC)
print("PrecisiÃ³n (Accuracy): {:.2f}%".format(accuracyConf*100))
```
